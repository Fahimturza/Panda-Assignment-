```python
import pandas as pd
from google.colab import files
uploaded = files.upload()
```

```python

data = pd.read_csv(next(iter(uploaded)), encoding='latin-1')

# 1. Get the dimensions of the dataset.
total_records, total_features = data.shape
print(f"1. Data Dimensions -> Records: {total_records}, Features: {total_features}")

# 2. Compute age statistics.
max_age = data['age'].max()
min_age = data['age'].min()
mean_age = data['age'].mean()
print(f"2. Age stats -> Maximum: {max_age}, Minimum: {min_age}, Average: {mean_age:.2f}")
# 3. Identify the top three most frequent full names.
data['full_name'] = data['first'] + " " + data['last']
top_names = data['full_name'].value_counts().head(3)
print("3. Top 3 common names:")
print(top_names)

# 4. Find customers with duplicate phone numbers.
dup_numbers = data['phone'].value_counts()[data['phone'].value_counts() > 1].index
shared_numbers = data[data['phone'].isin(dup_numbers)][['full_name', 'phone']]
print("4. Customers with shared phone numbers:")
print(shared_numbers.to_string(index=False))

# 5. Count how many customers are Structural Engineers.
structural_df = data[data['profession'] == "Structural Engineer"]
count_structural = structural_df.shape[0]
print(f"5. Structural Engineer count: {count_structural}")

# 6. Count male Structural Engineers.
male_structural = structural_df[structural_df['gender'] == "Male"].shape[0]
print(f"6. Male Structural Engineers: {male_structural}")

# 7. List female Structural Engineers from Alberta (province "AB").
female_ab = structural_df[(structural_df['gender'] == "Female") & (structural_df['province'] == "AB")]
print("7. Female Structural Engineers from Alberta:")
print(female_ab[['full_name', 'province']].to_string(index=False))

# 8. Determine spending stats.
highest_spend = data['price(CAD)'].max()
lowest_spend = data['price(CAD)'].min()
average_spend = data['price(CAD)'].mean()
print(f"8. Spending stats -> Highest: {highest_spend} CAD, Lowest: {lowest_spend} CAD, Average: {average_spend:.2f} CAD")

# 9. Find customers with zero spending.
no_spend = data[data['price(CAD)'] == 0][['full_name', 'price(CAD)']]
print("9. Customers with 0 CAD spending:")
print(no_spend.to_string(index=False))

# 10. Identify loyal customers (spent 100 CAD or more).
loyal_customers = data[data['price(CAD)'] >= 100][['full_name', 'price(CAD)']]
print("10. Loyal customers (spending >= 100 CAD):")
print(loyal_customers.to_string(index=False))

# 11. Count unique emails linked with a specific credit card.
unique_email_count = data[data['cc_no'] == '5020000000000230']['email'].nunique()
print(f"11. Unique emails for card 5020000000000230: {unique_email_count}")

# 12. Count cards expiring in 2019.
cards_2019 = data[data['cc_exp'].str.endswith("2019")]
print(f"12. Cards expiring in 2019: {cards_2019.shape[0]}")


    # 13. Determine how many customers use Visa.
    visa_users = data[data['cc_type'].str.contains("Visa", case=False, na=False)]
    print(f"13. Customers using Visa: {visa_users.shape[0]}")

    # 14. Find the customer(s) who spent exactly 100 CAD with a Visa card.
    visa_100 = data[(data['price(CAD)'] == 100) & (data['cc_type'].str.contains("Visa", case=False, na=False))]
    print("14. Customer(s) with exactly 100 CAD spent using Visa:")
    print(visa_100[['full_name', 'price(CAD)', 'cc_type']].to_string(index=False))

    # 15. Find the two most frequent professions.
    top_professions = data['profession'].value_counts().head(2)
    print("15. Top two professions:")
    print(top_professions)

    # 16. Extract and rank the top five email providers.
    data['email_provider'] = data['email'].apply(lambda x: x.split('@')[-1] if pd.notnull(x) else None)
    top_providers = data['email_provider'].value_counts().head(5)
    print("16. Top 5 email providers:")
    print(top_providers.to_dict())

    # 17. Check if any email contains "am.edu".
    amedu_found = data[data['email'].str.contains("am.edu", case=False, na=False)]
    if not amedu_found.empty:
        print("17. There is at least one email containing 'am.edu'.")
    else:
        print("17. No email contains 'am.edu'.")

    # 18. Determine which weekday has the highest customer count.
    busiest_day = data['weekday'].value_counts().idxmax()
    print(f"18. Busiest day of the week: {busiest_day}")
```
